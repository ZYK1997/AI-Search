```python
from collections import deque
```

```python
def BFS(maze, st, circleCheck = True):
    global n, m
    n, m = len(maze), len(maze[0])
    tot = 0
    Q = deque()
    S = set()
    Q.append((st, 0))
    S.add(st)
    while len(Q) > 0:
        x, dis = Q.popleft()
        tot += 1
        print(tot, x)
        if maze[x[0]][x[1]] == 'E':
            print("Find the END at {} with {} steps".format(x, dis))
            break
        for d in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            y = (x[0] + d[0], x[1] + d[1])
            if canVis(maze, y):
                if circleCheck:
                    if y not in S:
                        Q.append((y, dis + 1))
                        S.add(y)
                else:
                    Q.append((y, dis + 1))
```

```python
def bidirectional_search(maze, st, ed, circleCheck = True):
    global n, m
    n, m = len(maze), len(maze[0])
    tot = 0
    Q1 = deque()
    Q2 = deque()
    S1 = set()
    S2 = set()
    Q1.append((st, 0))
    Q2.append((ed, 0))
    S1.add(st)
    S2.add(ed)
    while len(Q1) > 0:
        x, d1 = Q1.popleft()
        y, d2 = Q2.popleft()
        tot += 2
        print(tot, x, y)
        if x in S2 or y in S1:
            print("Find the END with {} steps".format(d1 + d2))
            break
        for d in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            xx = (x[0] + d[0], x[1] + d[1])
            if canVis(maze, xx):
                if circleCheck:
                    if xx not in S1:
                        Q1.append((xx, d1 + 1))
                        S1.add(xx)
                else:
                    Q1.append((xx, d1 + 1))
            
            yy = (y[0] + d[0], y[1] + d[1])
            if canVis(maze, yy):
                if circleCheck:
                    if yy not in S2:
                        Q2.append((yy, d2 + 1))
                        S2.add(yy)
                else:
                    Q2.append((yy, d2 + 1))
```

```python
def IDF(maze, st, checkMode = 1):
    global n, m
    n, m = len(maze), len(maze[0])
    global tot, S
    tot = 0
    S = set()
    S.add(st)
    
    def dfs(x, dep):
        global tot, S
        tot += 1
        #print(tot, x)
        if maze[x[0]][x[1]] == 'E':
            print("Find the END at {}.".format(x))
            return True
        if dep == 0:
            return False
        for d in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            y = (x[0] + d[0], x[1] + d[1])
            if canVis(maze, y):
                if checkMode == 0:
                    if dfs(y, dep - 1):
                        return True
                elif checkMode == 1:
                    if y not in S:
                        S.add(y)
                        if dfs(y, dep - 1):
                            return True
                        S.remove(y)
                else:
                    print("checkMode error")
                    return False
        return False
    
    for dep in range(100):
        if dfs(st, dep):
            print("Minimum distance is {}.".format(dep))
            break
        else:
            print("With {} steps can't arrive the END.".format(dep))
    return tot
```

```python
def canVis(maze, y):
    global n, m
    if 0 <= y[0] and y[0] < n:
        if 0 <= y[1] and y[1] < m:
            return maze[y[0]][y[1]] != '0' and maze[y[0]][y[1]] != '%'
    return False
```

```python
def getMaze():
    with open("MazeData.txt") as f:
        ls = f.readlines()
        ls = list(map(lambda l: l.strip(), ls))
        maze1, maze2 = ls[:18], ls[21:]
        return maze1, maze2
```

```python
def findSymbol(maze, c):
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == c:
                return i, j
    return -1, -1

def findStart(maze):
    return findSymbol(maze, 'S')

def findEnd(maze):
    return findSymbol(maze, 'E')
```

```python
m1, m2 = getMaze()
st = findStart(m2)
BFS(m2, st)
```

```python
m1, m2 = getMaze()
st = findStart(m2)
IDF(m2, st)
```

```python
m1, m2 = getMaze()
st = findStart(m2)
ed = findEnd(m2)
bidirectional_search(m2, st, ed)
```
